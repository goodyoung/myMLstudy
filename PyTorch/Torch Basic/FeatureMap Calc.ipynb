## Feature Map 크기 계산 공식


- I: 입력 Feature Map의 크기
- F: Filter의 크기(Kernel size): 4*4
- P: Padding(정수): 모서리 채우는 것, 1이면 4*4 -> 5*5
- S: Stride(정수): 몇단위로 이동하는지

## Stride가 1이고 Padding이 없는 경우
- (5-3+0) / 1 + 1 = 3

```python
model1 = TestSimpleModel(1,1,3,1,0)
summary(model1, input_size=(1, 5, 5))  # (channels, height, width)
```

## Stride가 1이고 Padding이 1인 경우
- (5-3+2) / 1 + 1 = 5

```python
model2 = TestSimpleModel(1,1,3,1,1)
summary(model2, input_size=(1, 5, 5))  # (channels, height, width)
```

## Stride가 2이고 Padding이 없는 경우
- (5-3+0) / 2 + 1 = 2

```python
# (in_channels, out_channels, kernel_size, stride, padding)
model2 = TestSimpleModel(1,1,3,2,0)  
summary(model2, input_size=(1, 5, 5))  # (channels, height, width)
```

## Stride가 2이고 Padding이 1인 경우
- (5-3+2) / 2 + 1 = 3

```python
# (in_channels, out_channels, kernel_size, stride, padding)
model2 = TestSimpleModel(1,1,3,2,1)  
summary(model2, input_size=(1, 5, 5))  # (channels, height, width)
```
